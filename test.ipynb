```python
# Filter to only events starting with "A_"
a_events = df[df['concept:name'].str.startswith('A_')]

# For each case, get the last 'A_' activity
last_a = a_events.groupby('case:concept:name')['concept:name'].last().reset_index()
last_a.rename(columns={'concept:name': 'last_A_event'}, inplace=True)

# Merge back into the original dataframe
df = df.merge(last_a, on='case:concept:name', how='left')
df
```

```python
counts = (
    df[['case:concept:name', 'last_A_event']]
    .drop_duplicates(subset='case:concept:name')  # keep only one row per case
    ['last_A_event']
    .value_counts(dropna=False)
)
print(counts)
```

```python
# https://ais.win.tue.nl/bpi/2012/bautista.pdf 

df['case_declined'] = np.where(df['last_A_event'] == 'A_DECLINED', 1, 0)
df['case_cancelled'] = np.where(df['last_A_event'] == 'A_CANCELLED', 1, 0)
df['case_approved'] = np.where(df['last_A_event'].isin(['A_ACTIVATED', 'A_REGISTERED', 'A_APPROVED']), 1, 0)
df['case_unresolved'] = np.where(df['last_A_event'].isin(['A_FINALIZED', 'A_PREACCEPTED', 'A_ACCEPTED']), 1, 0)
```
